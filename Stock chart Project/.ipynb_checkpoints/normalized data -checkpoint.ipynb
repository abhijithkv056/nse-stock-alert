```python
df1 = df[['Date','Close']]
df1.info()
```

```python
c1 = []
for i in range(1,20):
    df['temp'] = df1['Close'] - df1['Close'].shift(-1*i)
    df['temp'] = df['temp']/df['Close']*100
    c1.append(df['temp'].to_numpy())

print(c1)    
df2 = pd.DataFrame(c1)
df3 = df2.transpose()
df3.head(10)
```

```python
df3.shape
print(len(c1))
```

```python
c2 =df3.to_numpy()
c2

```

```python
c3 =[]

for i in range(int(len(c2)-20)):
      
    c6 = np.concatenate((c2[i:i+20]))
    
    c3.append(c6)
    
                
```
