```python
import time
from nsetools import Nse
import yfinance as yf
import datetime
import os.path
import pandas as pd

# Create an instance of the NSE object
nse = Nse()

# Get the list of NSE stock symbols
nse_symbols = list(nse.get_stock_codes().keys())

# Set the start and end dates
end_date = datetime.date.today()
start_date = end_date - datetime.timedelta(days=30)

# Create an empty dictionary to store the data for each stock
data = {}

# Set the batch size to limit the number of stocks fetched at a time
batch_size = 50

# Set the delay between requests to avoid exceeding rate limits
delay = 5

# Loop through the list of stock symbols in batches and fetch their data
for i in range(0, len(nse_symbols), batch_size):
    batch_symbols = nse_symbols[i:i+batch_size]
    for symbol in batch_symbols:
        # Check if the data for the stock is already cached
        cache_file = f"{symbol}.csv"
        if os.path.isfile(cache_file):
            stock_data = pd.read_csv(cache_file, index_col='Date')
        else:
            stock_data = yf.download(symbol + ".NS", start=start_date, end=end_date)
            # Cache the fetched data
            stock_data.to_csv(cache_file)
        if not stock_data.empty:
            data[symbol] = stock_data
            print(f"Fetched {symbol}")
        else:
            print(f"No data found for {symbol}")
        # Add a delay between requests to avoid exceeding rate limits
        time.sleep(delay)

# Print the data for each stock
for symbol, stock_data in data.items():
    stock_name = nse.get_quote(symbol)['companyName']
    print(f"{stock_name} ({symbol}):")
    print(stock_data)
```
