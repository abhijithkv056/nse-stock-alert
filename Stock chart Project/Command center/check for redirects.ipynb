```python
from nsetools import Nse
import yfinance as yf
import datetime

# Create an instance of the NSE object
nse = Nse()

# Get the list of NSE stock symbols
nse_symbols = nse.get_stock_codes()

# Set the start and end dates
end_date = datetime.date.today()
start_date = end_date - datetime.timedelta(days=30)

# Create an empty dictionary to store the data for each stock
data = {}

# Loop through the list of stock symbols and fetch their data
for symbol in nse_symbols:
    stock_data = yf.download(symbol + ".NS", start=start_date, end=end_date)
    if not stock_data.empty:
        data[symbol] = stock_data

# Print the data for each stock
for symbol, stock_data in data.items():
    stock_name = nse_symbols[symbol]
    print(f"{stock_name} ({symbol}):")
    print(stock_data)
```
